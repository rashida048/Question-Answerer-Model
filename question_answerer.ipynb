```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
context = "Rashida is a dancer. She started dance school when she was 5. All her life dedicated to only art and dance. She grew in Brazil in a coffee firm. Rashida never went to school in her life. All her life she only went to coffee firms. Today she drinks 12 cups of coffee every mroning."
```

```python
result = question_answerer(question="Where did Rashida grow up?", context = context)
print(f"answer: '{result['answer']}'")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=700, chunk_overlap=0)
texts = text_splitter.split_text(test_on_test)
```

```python
question_answerer = pipeline("question-answering")
context = texts
```

```python
result = question_answerer(question="Which datasets are good Data Analysis?", context = test_on_test)
print(f"answer: '{result['answer']}'")
```
